```python
import numpy as np
from numpy import linalg as LA
import random
import math
import matplotlib.pyplot as plt
```

```python
matrix = [
    [3, 4, -2],
    [-2, 8, 1],
    [0, 15, -7]
]

matrix = np.matrix(matrix)
```

```python
# inverse of the matrix

print("Inverse: \n" + str(LA.inv(matrix)))

# determinant of matrix

print("Determinant: \n" + str(LA.det(matrix)))

# eigenvalues of the matrix

print("Eigenvalues: \n" + str(LA.eig(matrix)[0]))

# eigenvectors

print("Eigenvectors: \n" + str(LA.eig(matrix)[1]))

def genRand(amt, max = 5):
    output = []
    for i in range(amt):
        output.append(random.randint(0, max))
    return output

randNums = genRand(10)

print(randNums)
```

```python
def gaussian_pdf(args, mean = 2, var = 1):
    sd = math.sqrt(var)
    output = []
    for i in args:
        den = sd * math.sqrt(2 * math.pi)
        exp = (-1/2) * (((i - mean) / sd) ** 2)
        output.append(math.exp(exp) / den)
    return output

print(gaussian_pdf(randNums))
```

```python
randNums = (1, 2, 1.5, 2.5, 0.6, 3.6)

print(gaussian_pdf(randNums))
```

```python
def genGaussianRand(amt, mean, var):
    sd = math.sqrt(var)
    return np.random.normal(mean, sd, amt)

print(genGaussianRand(10, 4, 8))
```

```python
arr = []
for i in range(-20, 21):
    arr.append(i)

def poly(x):
    return (x ** 2) - ( 4 * x ) + 5

print(arr)

newArr = []
for i in arr:
    newArr.append(poly(i))
print(newArr)
```

```python
plt.plot(arr, newArr)
plt.ylabel('Y')
plt.xlabel('X')
plt.title('y = x^2 - 4x + 5')
plt.show()
```
